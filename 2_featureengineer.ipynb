```python
import pandas as pd
```

```python
data_cleaned = pd.read_csv("post_processed_data_cleaned.csv")
data_cleaned.head()
```

```python
data_cleaned['video_view_count'].replace(0, pd.NA, inplace=True)
data_cleaned.head()
```

```python
data_cleaned['engagement_rate'] = (
    (data_cleaned['video_like_count'] + data_cleaned['video_comment_count']) /
    data_cleaned['video_view_count']
)
data_cleaned['engagement_rate'].head()
```

```python
import pandas as pd

data_cleaned['video_trending__date'] = pd.to_datetime(data_cleaned['video_trending__date'])

data_cleaned['video_published_at'] = pd.to_datetime(data_cleaned['video_published_at'])

data_cleaned['video_published_at'] = data_cleaned['video_published_at'].dt.tz_localize(None)
data_cleaned['time_difference'] = data_cleaned['video_trending__date'] - data_cleaned['video_published_at']
data_cleaned[['video_trending__date', 'video_published_at', 'time_difference']]
```

```python
data_cleaned['video_duration']
```

```python
import pandas as pd
import isodate

data_cleaned['video_duration_seconds'] = data_cleaned['video_duration'].apply(lambda x: isodate.parse_duration(x).total_seconds())
```

```python
feature_table_fixed = data_cleaned[['engagement_rate', 'time_difference', 'video_duration_seconds']].head()
feature_table_fixed
```
